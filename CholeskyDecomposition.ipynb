```python
import math
import numpy as np
```

```python
def matrix_check_spd(matA):
  check1=np.array_equal(matA,matA.T)
  if check1:
    check2=np.all(np.linalg.eigvals(matA)>0)
    if check2:
      return True

  return False
```

```python
def cholesky(A):
    n = len(A)
    L = [[0]*n for _ in range(n)]

    for i in range(n):
        for j in range(i+1):
            s = sum(L[i][k] * L[j][k] for k in range(j))

            if i == j:
                L[i][j] = math.sqrt(A[i][i] - s)
            else:
                L[i][j] = (A[i][j] - s) / L[j][j]

    return np.array(L)
```
